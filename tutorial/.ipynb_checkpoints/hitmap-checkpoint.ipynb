```julia
using Falcons

using PyCall
using PyPlot
hp = pyimport("healpy")
```

```julia
day = 60 * 60 * 24
year = day * 365

#= Scan strategy parameter setting=#
ss = scan_strategy()
ss.nside = 128
ss.times = year
ss.sampling_rate = 1
ss.FP_theta = [0.0]
ss.FP_phi = [0.0]
ss.alpha = 55
ss.beta = 60
ss.prec_period = 180.22
ss.spin_rpm = 0.04
ss.hwp_rpm = 0.05
ss.start_point = "pole" #You can choose "pole" or "equator"
```

```julia
pix_tod, psi_tod = get_scan_tod(ss, 0, 10000)
temp_map_bore = HitMap(ss.nside, pix_tod)
hp.mollview(temp_map_bore, rot=(0,0), cmap="viridis")
hp.graticule()
```
